- 使用 標籤 「類別名稱」取得資料
網頁是由**標籤**所組成，基本結構如下
```html
<標籤名稱 class="類別名稱">內容</標籤名稱>
<標籤名稱 id="id名稱">內容</標籤名稱>
```
今天我們要擷取的任何內容，一定是被包裹在在某一個標籤裡面

要擷取資料前，首先需要透過方法**選擇**到該標籤

這時我們就需要介紹一下在網頁開發裡的強大套件：**jQuery**

## jQuery是什麼？
網頁元素取得神器，使用css的選擇器選取元素，只要會css的選法，就能快速找到想要的元素
```javascript

```
PyQuery 使用類似 jQuery的方式取得資料


# 即時匯率監控


````python
import xlwings as xw
from pyquery import PyQuery as pq
from xlwings.constants import Direction
import datetime as dt
import time

WB = xw.Book(r'C:\Users\user\Desktop\L4\currency_trading.xlsm')
ROW = 1

def crawler():
    url = "http://rate.bot.com.tw/xrt?Lang=zh-TW"
    selector = "td[class^='rate-content-']"
    # 抓取台銀的匯率頁面，存入html物件
    html = pq(url)
    html_elements = html(selector)

    return html_elements 

# 定義 clean_up 函式
def clean_up(html_elements):
    # 將傳入的html_elements轉為文字
    html_elements_text = html_elements.text();
    # 建立幣別清單
    print(html_elements_text)
    currency_list = ["USD", "HKD", "GBP", "AUD", "CAD", "SGD", "CHF", "JPY", "ZAR", "SEK", "NZD", "THB", "PHP", "IDR", "EUR", "KRW", "VND", "MYR", "CNY"]
    # 分割轉好的文字，產生匯率的清單
    rates_list = html_elements_text.split(' ')
    # 建立空的字典檔，當成資料集
    data_set = {}
    # 對於每一個在 幣別清單中的 index
    for index in range(len(currency_list)):
        # 取得 幣別的名稱
        currency_name = currency_list[index]
        # 根據幣別清單的index 計算 匯率清單的位置 並將相關數據存在變數中備用
        cash_buy   = rates_list[index * 4]
        cash_sell  = rates_list[index * 4 + 1]
        sight_buy  = rates_list[index * 4 + 2]
        sight_sell = rates_list[index * 4 + 3]
        # 在字典中，新增一個 key， 內容為 另一個字典 負責放四種匯率值
        data_set[currency_name] = {
            'CASH_BUY'  : cash_buy,
            'CASH_SELL' : cash_sell,
            'SIGHT_BUY' : sight_buy,
            'SIGHT_SELL': sight_sell
        }
        # 回傳 建立好的資料集

    return data_set

def record_data(currency_sheet, currency_data, row):
    print("row = {}".format(row))
    currency_sheet.cells(row, 1).number_format = 'yyyy/mm/dd hh:mm:ss'
    currency_sheet.cells(row, 1).value = dt.datetime.now()
    currency_sheet.cells(row, 2).value = currency_data['CASH_BUY']
    currency_sheet.cells(row, 3).value = currency_data['CASH_SELL']
    currency_sheet.cells(row, 4).value = currency_data['SIGHT_BUY']
    currency_sheet.cells(row, 5).value = currency_data['SIGHT_SELL']

def write_excel(workbook, data_set, sheet_row):
    sheet = workbook.sheets['Dashboard']
    # 偵測目前總共有幾筆資料
    num_of_rows = sheet.range('A1').end(Direction.xlDown).row
    # 到 dashboard 裏，從第二個 row 開始，找出所有關注的貨幣
    sheet.range((2, 2), (num_of_rows, 5)).clear_contents()

    for row in range(2, num_of_rows+1):
        # 在 A 欄找到 貨幣名稱
        currency = sheet.cells(row, 1).value
        # 用貨幣名稱到 data_set 取出該貨幣所有資料
        currency_data = data_set[currency]
        # 將資料寫入相對應的欄位
        sheet.cells(row, 2).value = currency_data['CASH_BUY']
        sheet.cells(row, 3).value = currency_data['CASH_SELL']
        sheet.cells(row, 4).value = currency_data['SIGHT_BUY']
        sheet.cells(row, 5).value = currency_data['SIGHT_SELL']
        currency_sheet = workbook.sheets[currency]
        currency_sheet.range("A1").value = "時間"
        record_data(currency_sheet, currency_data, sheet_row)    

def write_time(workbook):
    sheet = workbook.sheets['Dashboard']
    # 設定時間顯示格式
    sheet.cells(2, 7).number_format = 'yyyy/mm/dd hh:mm:ss'
    # 將現在的時間寫入 Excel
    sheet.cells(2, 7).value = dt.datetime.now()

""" 主程式從這裏開始 """
# 用 while 寫成的無窮回圈 
while True:
    # 呼叫爬蟲截取匯率資料
    html_elements = crawler()
    # 使用 clean_up 函式清理資料
    data_set = clean_up(html_elements)
    # 列印清理好的資料集
    # print(data_set)
    # 將資料寫入 excel
    ROW = ROW + 1
    write_excel(WB, data_set, ROW)
    # 將時間寫入 excel
    write_time(WB)
    # 每隔 5 秒執行一次
    time.sleep(5)
````
